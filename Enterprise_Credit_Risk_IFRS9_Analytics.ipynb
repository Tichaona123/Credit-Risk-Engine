## 1.1 Credit Risk Business Context

Credit risk, also known as default risk, represents the possibility that a borrower will fail to meet their contractual obligations regarding loan repayment. For financial institutions, effective credit risk management is not merely a competitive advantage---it is a regulatory imperative and a fundamental pillar of financial stability.

### The Lending Decision Process

When a borrower applies for a loan, the financial institution must evaluate multiple dimensions of risk:

- **Borrower Character:** Credit history, employment stability, income adequacy
- **Capacity to Repay:** Debt-to-income ratios, payment burden analysis
- **Capital:** Collateral quality, loan-to-value ratios
- **Conditions:** Economic environment, sector-specific risks
- **Collateral:** Security backing the loan facility

This evaluation, often referred to as the "5 Cs of Credit," forms the foundation of traditional credit analysis. Modern credit risk modeling augments this framework with advanced statistical and machine learning techniques to achieve greater accuracy and consistency.

## 1.5 Credit Risk Analytics Foundations

### Probability of Default (PD)

Probability of Default represents the likelihood that a borrower will default on their obligations over a specified time horizon, typically 12 months. PD is a fundamental input to:

- **Expected Credit Loss (ECL)** calculations under IFRS 9
- **Regulatory Capital** requirements under Basel III/IV
- **Risk-Based Pricing** strategies
- **Portfolio Monitoring** and early warning systems

The mathematical formulation of PD estimation through logistic regression is:

$$\text{PD} = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \ldots + \beta_n X_n)}}$$

Where $\beta_0$ is the intercept, $\beta_i$ are the coefficients, and $X_i$ are the risk factors.

### Credit Scoring Interpretation

Credit scores (probability of default estimates) are typically mapped to risk grades or score bands:

| Score Band | Risk Grade | PD Range | Interpretation |
|------------|-----------|----------|----------------|
| 0.00 - 0.05 | Very Low | 0% - 5% | Excellent creditworthiness |
| 0.05 - 0.15 | Low | 5% - 15% | Good credit quality |
| 0.15 - 0.30 | Medium | 15% - 30% | Moderate risk, enhanced monitoring |
| 0.30 - 0.50 | High | 30% - 50% | Elevated risk, close watch |
| 0.50 - 1.00 | Very High | 50% - 100% | Critical risk, intervention required |

### The Cost of Misclassification

In credit risk, not all prediction errors have equal consequences:

- **False Negative (Type II Error):** Predicting a loan as safe when it defaults
  - **Cost:** Loss of principal and interest, collection expenses, legal fees
  - **Typical Impact:** 3-5x the loan amount in total losses

- **False Positive (Type I Error):** Predicting default when the loan would perform
  - **Cost:** Lost interest income, customer relationship damage
  - **Typical Impact:** 0.5-1x the loan's net present value

This asymmetric cost structure explains why **Recall** (sensitivity) is often prioritized in default prediction models, and why the **F1-Score** at optimized thresholds serves as a key business metric.

# 2. IFRS 9 Regulatory Framework

## 2.1 Introduction to IFRS 9

IFRS 9 (International Financial Reporting Standard 9) is the accounting standard issued by the International Accounting Standards Board (IASB) that governs the classification, measurement, and impairment of financial instruments. It replaced IAS 39 in January 2018 and introduced a forward-looking **Expected Credit Loss (ECL)** model that fundamentally changed how financial institutions account for credit impairment.

### Key Differences: Incurred Loss vs. Expected Loss

| Aspect | IAS 39 (Incurred Loss) | IFRS 9 (Expected Loss) |
|--------|----------------------|----------------------|
| Timing of Recognition | After loss event occurs | Before loss event occurs |
| Economic Signal | Backward-looking | Forward-looking |
| Provision Volatility | Sharp increases during crises | Smoother, anticipatory build-up |
| Capital Impact | Procyclical | Less procyclical |
| Data Requirements | Historical loss data | PD, LGD, EAD projections |

The shift from incurred loss to expected loss represents one of the most significant changes in banking accounting in decades, requiring institutions to develop sophisticated predictive models.

## 2.2 Expected Credit Loss (ECL) Framework

The fundamental equation for Expected Credit Loss under IFRS 9 is:

$$\text{ECL} = \text{PD} \times \text{LGD} \times \text{EAD}$$

Where:

- **PD (Probability of Default):** The likelihood of default over a specified time horizon
- **LGD (Loss Given Default):** The percentage of exposure that will not be recovered after default
- **EAD (Exposure at Default):** The outstanding loan balance at the time of default

### Lifetime ECL vs. 12-Month ECL

IFRS 9 requires different measurement approaches based on credit risk staging:

| Measure | Formula | When Applied |
|---------|---------|--------------|
| 12-Month ECL | PD(12m) x LGD x EAD | Stage 1: Performing assets |
| Lifetime ECL | PD(Lifetime) x LGD x EAD | Stage 2: Significant risk increase |
| Lifetime ECL | PD(Lifetime) x LGD x EAD | Stage 3: Credit-impaired assets |

## 2.3 IFRS 9 Staging Assessment

The staging determination is the critical first step in ECL measurement:

### Stage 1: Performing Assets

- **Definition:** Financial instruments with no significant increase in credit risk since origination
- **ECL Measurement:** 12-month expected credit losses
- **Interest Recognition:** On gross carrying amount
- **Portfolio Composition:** Typically 70-85% of total portfolio

### Stage 2: Significant Increase in Credit Risk (SICR)

- **Definition:** Credit risk has increased significantly since initial recognition
- **ECL Measurement:** Lifetime expected credit losses
- **Interest Recognition:** On gross carrying amount
- **SICR Triggers:**
  - More than 30 days past due (rebuttable presumption)
  - Significant deterioration in credit rating
  - Material decline in collateral value
  - Borrower placed on internal watchlist
  - Significant adverse change in borrower's financial condition

### Stage 3: Credit-Impaired Assets

- **Definition:** Objective evidence of impairment at the reporting date
- **ECL Measurement:** Lifetime expected credit losses
- **Interest Recognition:** On net carrying amount (amortized cost less provision)
- **Indicators:**
  - Significant financial difficulty of the borrower
  - Breach of contract (90+ days past due)
  - Restructuring with concessionary terms
  - Probability of bankruptcy or liquidation

### Stage Transition Diagram

The migration between stages follows the credit deterioration pathway:

**Stage 1 (Performing) ---SICR Trigger--> Stage 2 (Deteriorated) ---Default Event--> Stage 3 (Impaired)**

## 2.4 Regulatory Importance of Explainable Models

IFRS 9 and banking regulators place significant emphasis on model transparency:

### European Banking Authority (EBA) Guidelines

The EBA Guidelines on PD estimation, LGD estimation, and treatment of defaulted assets (EBA/GL/2017/16) require:

1. **Model Documentation:** Comprehensive documentation of model development, validation, and monitoring
2. **Data Quality:** Rigorous data quality assessment and governance
3. **Statistical Rigor:** Appropriate statistical methods with demonstrated performance
4. **Expert Judgment:** Transparent incorporation of expert judgment where statistical data is insufficient
5. **Regular Validation:** Annual independent validation of model performance

### Why Explainability Matters in Banking

| Stakeholder | Explainability Need |
|-------------|-------------------|
| External Auditors | Verify model appropriateness and parameter reasonableness |
| Banking Regulators | Assess compliance with prudential requirements |
| Internal Validation | Independent review and challenger model development |
| Credit Committees | Understand drivers of portfolio risk trends |
| Board of Directors | Oversight of model risk and material model changes |

This project addresses these requirements through comprehensive SHAP-based explainability analysis, providing both global (portfolio-level) and local (individual loan-level) feature attribution.

## 2.5 Basel and IFRS 9 Relationship

While developed independently, Basel regulatory capital frameworks and IFRS 9 accounting standards share common foundations:

| Dimension | Basel III/IV | IFRS 9 |
|-----------|-------------|--------|
| Purpose | Regulatory capital | Financial reporting |
| PD Horizon | 12-month through-the-cycle | 12-month or lifetime point-in-time |
| LGD | Downturn LGD | Unbiased, probability-weighted |
| EAD | Effective maturity-adjusted | Expected at default date |
| Overlays | Margin of conservatism | Unbiased, forward-looking |

The model developed in this project provides **point-in-time PD estimates** suitable for IFRS 9 ECL calculations, which can be further adjusted for through-the-cycle parameters for Basel capital purposes.

## 4.2 Temporal Features

Temporal features capture the structural characteristics of the loan and potential seasonality effects. Loan duration reflects the extended risk exposure period, while approval timing may correlate with economic cycles or institutional lending patterns.

**Key Temporal Features:**
- `loan_duration_days`: Total days from disbursement to maturity---longer loans have extended exposure
- `days_approval_to_disburse`: Processing efficiency proxy; delays may indicate manual review for riskier applications
- `approval_month/quarter`: Seasonal effects on agricultural and business lending
- `age_bucket`: Life-stage risk segmentation (younger borrowers often exhibit higher default rates)

## 4.3 Financial Features

Financial ratio features are the most predictive in credit risk modeling. The **Debt-to-Income (DTI) ratio** is universally recognized as the single most powerful predictor of consumer loan default.

**DTI Ratio Formula:**

$$\text{DTI Ratio} = \frac{\text{Estimated Monthly Payment}}{\text{Monthly Income}}$$

**Interpretation Guidelines:**

| DTI Range | Risk Level | Lending Action |
|-----------|-----------|----------------|
| < 20% | Low | Standard approval |
| 20% - 36% | Moderate | Standard approval with documentation |
| 36% - 50% | Elevated | Enhanced review, possible conditions |
| > 50% | High | Decline or exception approval required |

The estimated monthly payment is calculated using the standard loan amortization formula:

$$\text{PMT} = P \times \frac{r(1+r)^n}{(1+r)^n - 1}$$

Where $P$ is the principal, $r$ is the monthly interest rate, and $n$ is the number of payments.

## 4.4 Risk Flags

Binary risk indicators provide intuitive, rule-based signals that complement continuous model scores:

| Flag | Condition | Risk Logic |
|------|-----------|------------|
| is_mfi_loan | Rate > 30% | Microfinance loans carry higher structural risk |
| is_high_rate | Rate > 100% | Extreme interest rates indicate subprime borrowers |
| high_dti | DTI > 50% | Severe payment burden relative to income |
| very_new_employee | Tenure <= 3 months | Employment instability increases default risk |
| high_obligations | 3+ existing loans | Over-leveraged borrower profile |

## 4.5 Interaction Features

Interaction features capture combined effects that individual variables cannot express. For example, a high interest rate combined with a long term creates compounding payment pressure that exceeds the sum of individual risks.

## 4.6 Target Encoding with Out-of-Fold Validation

High-cardinality categorical variables (province, product_code) are encoded using target statistics with out-of-fold (OOF) validation to prevent information leakage. The smoothing formula prevents overfitting to rare categories:

$$\text{Encoded Value} = \frac{n_{cat} \times \bar{y}_{cat} + \alpha \times \bar{y}_{global}}{n_{cat} + \alpha}$$

```python
# ============================================================
# TEMPORAL FEATURES (Clean, no leakage)
# ============================================================
# Loan lifecycle
df['loan_duration_days'] = (df['maturity_date'] - df['date_disbursed']).dt.days
df['days_approval_to_disburse'] = (df['date_disbursed'] - df['date_approved']).dt.days
df['days_to_first_payment'] = (df['first_payment_due'] - df['date_disbursed']).dt.days

# Seasonality
df['approval_month'] = df['date_approved'].dt.month
df['approval_quarter'] = df['date_approved'].dt.quarter
df['approval_dayofweek'] = df['date_approved'].dt.dayofweek

# Client age buckets
df['age_bucket'] = pd.cut(df['client_age'],
                          bins=[0, 25, 35, 45, 55, 100],
                          labels=[0, 1, 2, 3, 4]).astype(float)

print("Temporal features created.")

# ============================================================
# FINANCIAL FEATURES (Core credit risk metrics)
# ============================================================
# Monthly interest rate
monthly_rate = df['annual_rate_pct'].fillna(df['annual_rate_pct'].median()) / 100 / 12

# Estimated monthly payment (amortization formula)
n = df['term_months']
df['est_monthly_payment'] = df['amount_usd'] * (
    monthly_rate * (1 + monthly_rate)**n
) / ((1 + monthly_rate)**n - 1)

# Fix edge cases
low_rate = monthly_rate < 0.001
df.loc[low_rate, 'est_monthly_payment'] = df.loc[low_rate, 'amount_usd'] / df.loc[low_rate, 'term_months']

# Debt-to-Income ratio (THE most important credit risk metric)
income = df['monthly_income_usd'].copy()
df['dti_ratio'] = df['est_monthly_payment'] / income.replace(0, np.nan)
df['dti_ratio'] = df['dti_ratio'].clip(0, 5)  # Cap at 500%

# Total cost of loan
df['total_loan_cost'] = df['est_monthly_payment'] * df['term_months']
df['interest_to_principal'] = (df['total_loan_cost'] - df['amount_usd']) / df['amount_usd'].replace(0, np.nan)

# Income ratios
df['income_to_loan'] = income / df['amount_usd'].replace(0, np.nan)
df['loan_to_annual_income'] = df['amount_usd'] / (income * 12).replace(0, np.nan)

# Amount per month (payment intensity)
df['amount_per_term_month'] = df['amount_usd'] / df['term_months']

# Log transforms (reduce skewness)
df['log_amount'] = np.log1p(df['amount_usd'])
df['log_income'] = np.log1p(income.fillna(0))
df['log_rate'] = np.log1p(df['annual_rate_pct'].fillna(0))

print("Financial features created.")

# ============================================================
# RISK FLAGS (Binary indicators)
# ============================================================
# Loan type segmentation
df['is_mfi_loan'] = (df['annual_rate_pct'].fillna(0) > 30).astype(int)
df['is_high_rate'] = (df['annual_rate_pct'].fillna(0) > 100).astype(int)
df['is_short_term'] = (df['term_months'] <= 6).astype(int)
df['is_long_term'] = (df['term_months'] >= 24).astype(int)

# Collateral
df['has_real_collateral'] = df['collateral_type'].isin(['Vehicle', 'Property', 'Savings']).astype(int)
df['has_any_collateral'] = (~df['collateral_type'].isin(['None']) & df['collateral_type'].notna()).astype(int)

# Employment stability
df['stable_employment'] = (df['months_at_employer'].fillna(0) >= 24).astype(int)
df['very_new_employee'] = (df['months_at_employer'].fillna(0) <= 3).astype(int)

# Multiple obligations
df['high_obligations'] = (df['existing_obligations'] >= 3).astype(int)

# Age risk
df['is_young'] = (df['client_age'] < 25).astype(int)

# High DTI
df['high_dti'] = (df['dti_ratio'].fillna(0) > 0.5).astype(int)

print("Risk flags created.")

# ============================================================
# INTERACTION FEATURES (Careful, only strong ones)
# ============================================================
df['rate_x_term'] = df['annual_rate_pct'].fillna(0) * df['term_months']
df['amount_x_rate'] = df['amount_usd'] * df['annual_rate_pct'].fillna(0) / 1000

# Obligations burden
df['obligations_ratio'] = df['existing_obligations'] / (df['num_dependents'].fillna(0) + 1)

# Categorical Cross Features
df['province_x_sector'] = df['province'].astype(str) + "_" + df['employment_sector'].astype(str)
df['purpose_x_product'] = df['loan_purpose'].astype(str) + "_" + df['product_code'].astype(str)

print("Interaction features created.")

# ============================================================
# MISSINGNESS INDICATORS (Only for key columns)
# ============================================================
important_missing = ['monthly_income_usd', 'collateral_type', 'months_at_employer',
                     'employment_sector', 'annual_rate_pct']
for col in important_missing:
    df[f'miss_{col}'] = df[col].isnull().astype(int)

df['total_missing'] = df[[f'miss_{c}' for c in important_missing]].sum(axis=1)

print(f"Missingness indicators created ({len(important_missing)} columns).")

# ============================================================
# CATEGORICAL ENCODING (For CatBoost native + others)
# ============================================================
# Define categorical columns for CatBoost
cat_features_raw = ['product_code', 'payment_frequency', 'loan_purpose',
                    'client_gender', 'marital_status', 'employment_sector',
                    'collateral_type', 'disbursement_channel', 'province']

# Frequency encoding (safe, no target leakage)
for col in cat_features_raw:
    freq = df[col].value_counts(normalize=True)
    df[f'{col}_freq'] = df[col].map(freq).fillna(0)

# Payment frequency ordinal
freq_map = {'Weekly': 0, 'Bi-Weekly': 1, 'Monthly': 2}
df['payment_freq_ord'] = df['payment_frequency'].map(freq_map).fillna(1)

# Gender encoding
df['gender_enc'] = df['client_gender'].map({'Male': 0, 'Female': 1}).fillna(-1)

# For CatBoost: fill NaN in categoricals with 'Missing'
for col in cat_features_raw:
    df[col] = df[col].fillna('Missing').astype(str)

# Product code as string for CatBoost
df['product_code'] = df['product_code'].astype(str)
df['province_x_sector'] = df['province_x_sector'].fillna('Missing').astype(str)
df['purpose_x_product'] = df['purpose_x_product'].fillna('Missing').astype(str)
cat_features_raw.extend(['province_x_sector', 'purpose_x_product'])

print("Categorical encoding complete.")
```

```python
# ============================================================
# BUILD FEATURE MATRIX
# ============================================================
# Numeric features (no target-leaky aggregations!)
numeric_features = [
    # Raw numeric
    'amount_usd', 'annual_rate_pct', 'term_months', 'monthly_income_usd',
    'existing_obligations', 'num_dependents', 'months_at_employer',
    # Age
    'client_age', 'age_bucket',
    # Temporal
    'loan_duration_days', 'days_approval_to_disburse', 'days_to_first_payment',
    'approval_month', 'approval_quarter', 'approval_dayofweek',
    # Financial ratios
    'est_monthly_payment', 'dti_ratio', 'total_loan_cost',
    'interest_to_principal', 'income_to_loan', 'loan_to_annual_income',
    'amount_per_term_month',
    # Log transforms
    'log_amount', 'log_income', 'log_rate',
    # Risk flags
    'is_mfi_loan', 'is_high_rate', 'is_short_term', 'is_long_term',
    'has_real_collateral', 'has_any_collateral',
    'stable_employment', 'very_new_employee',
    'high_obligations', 'is_young', 'high_dti',
    # Interactions
    'rate_x_term', 'amount_x_rate', 'obligations_ratio',
    # Missingness
    'miss_monthly_income_usd', 'miss_collateral_type', 'miss_months_at_employer',
    'miss_employment_sector', 'miss_annual_rate_pct', 'total_missing',
    # Frequency encodings
    'payment_freq_ord', 'gender_enc',
]

# Add frequency encoded features
freq_features = [f'{col}_freq' for col in cat_features_raw]
numeric_features += freq_features

# CatBoost categorical features
catboost_cat_cols = cat_features_raw

# All features for non-CatBoost models (numeric only)
all_numeric_features = [f for f in numeric_features if f in df.columns]

print(f"Numeric features: {len(all_numeric_features)}")
print(f"CatBoost categorical features: {len(catboost_cat_cols)}")

# Split back
train_df = df[df['is_train'] == 1].copy()
test_df = df[df['is_train'] == 0].copy()

y = train_df[TARGET].astype(int)

# ============================================================
# OOF TARGET ENCODING (Smooth & Leak-Free)
# ============================================================
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=SEED)

te_features = []
# We'll encode high-cardinality and important categoricals
cols_to_encode = ['province', 'product_code', 'employment_sector', 'loan_purpose', 'province_x_sector']
for col in cols_to_encode:
    new_col = f'{col}_te'
    train_df[new_col] = 0.0
    test_df[new_col] = 0.0

    global_mean = y.mean()

    # OOF for train
    for tr_idx, val_idx in kf.split(train_df):
        X_tr, X_val = train_df.iloc[tr_idx], train_df.iloc[val_idx]
        # Calculate mean target per category with smoothing
        smooth = 10
        counts = X_tr.groupby(col)[TARGET].count()
        means = X_tr.groupby(col)[TARGET].mean()

        smoothed = (counts * means + smooth * global_mean) / (counts + smooth)
        train_df.loc[val_idx, new_col] = X_val[col].map(smoothed).fillna(global_mean)

    # Full train for test
    counts = train_df.groupby(col)[TARGET].count()
    means = train_df.groupby(col)[TARGET].mean()
    smoothed = (counts * means + smooth * global_mean) / (counts + smooth)
    test_df[new_col] = test_df[col].map(smoothed).fillna(global_mean)

    te_features.append(new_col)

numeric_features.extend(te_features)
all_numeric_features = [f for f in numeric_features if f in train_df.columns]
print(f"Added {len(te_features)} OOF target encoded features.")

# For CatBoost: use both numeric + categorical
X_cb = train_df[all_numeric_features + catboost_cat_cols].copy()
X_cb_test = test_df[all_numeric_features + catboost_cat_cols].copy()

# For LightGBM/XGBoost: numeric only
X_num = train_df[all_numeric_features].copy().replace([np.inf, -np.inf], np.nan)
X_num_test = test_df[all_numeric_features].copy().replace([np.inf, -np.inf], np.nan)

test_ids = test_df[ID_COL].copy()

print(f"\nX_catboost: {X_cb.shape}")
print(f"X_numeric:  {X_num.shape}")
print(f"y: {y.shape} (default rate: {y.mean():.2%})")
```

# 5. Adversarial Validation

## 5.1 Concept and Business Rationale

Adversarial validation is a critical diagnostic technique in machine learning model development, particularly for credit risk applications where model generalization directly impacts financial outcomes. The method tests whether the training and test datasets originate from the same underlying distribution by training a classifier to distinguish between them.

### Why This Matters for Banking

In production credit risk models, distribution shift between training and deployment data is one of the most common causes of model degradation. If the training data (historical loans) differs systematically from the data the model will encounter in production (new applications), the model's predictions may be unreliable regardless of its cross-validation performance.

### Interpretation of Results

| Adversarial AUC | Interpretation | Action Required |
|-----------------|----------------|-----------------|
| 0.50 | Perfect distribution match | None---proceed with confidence |
| 0.50 - 0.55 | Minor differences, likely acceptable | Monitor; investigate top features |
| 0.55 - 0.60 | Moderate shift; requires attention | Remove/re-engineer discriminative features |
| > 0.60 | Significant distribution shift | Major feature revision required |

### Methodology

The adversarial validation process:
1. Concatenate training and test feature matrices
2. Create a binary target: 1 for training samples, 0 for test samples
3. Train a Random Forest classifier to distinguish between them
4. Evaluate cross-validated AUC
5. Examine feature importances to identify problematic variables

Features with high adversarial importance indicate variables whose distributions differ between train and test. These features may cause the model to learn train-specific patterns rather than generalizable risk relationships.

**Statistical Foundation:** This technique is grounded in the concept of **covariate shift**, where $P_{train}(X) \neq P_{test}(X)$ while $P(Y|X)$ remains constant. Under covariate shift, models that rely on shifted features generalize poorly to the test distribution.


```python
# ============================================================
# ADVERSARIAL VALIDATION
# ============================================================
# Can a model distinguish train from test?
from sklearn.ensemble import RandomForestClassifier

adv_X = pd.concat([X_num, X_num_test], axis=0)
adv_y = np.concatenate([np.ones(len(X_num)), np.zeros(len(X_num_test))])

# Quick RF to check
adv_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=SEED, n_jobs=-1)
adv_scores = cross_val_score(adv_model, adv_X.fillna(-999), adv_y,
                             cv=3, scoring='roc_auc')
print(f"Adversarial Validation AUC: {adv_scores.mean():.4f}")
print(f"  (0.50 = train/test identical, >0.60 = significant distribution shift)")

# Feature importance from adversarial model
adv_model.fit(adv_X.fillna(-999), adv_y)
adv_importance = pd.Series(adv_model.feature_importances_,
                           index=all_numeric_features).sort_values(ascending=False)
print(f"\nTop features distinguishing train from test (potential problems):")
print(adv_importance.head(10).to_string())

# Features with very high adversarial importance are suspicious
suspicious_features = adv_importance[adv_importance > 0.05].index.tolist()
if suspicious_features:
    print(f"\n⚠️  Potentially problematic features: {suspicious_features}")
    print("   Consider removing these if they don't improve CV score.")
```

# 6. Model Training

## 6.1 Algorithm Selection Rationale

This solution implements a multi-algorithm ensemble strategy using three leading gradient boosting frameworks. Each algorithm brings distinct advantages that contribute to ensemble diversity and robustness.

### CatBoost (Primary Model)

CatBoost, developed by Yandex, serves as the primary model due to several unique advantages for credit risk modeling:

**Native Categorical Handling:** CatBoost uses ordered target statistics for categorical features, eliminating the need for manual encoding. This preserves categorical information without introducing the ordinality assumptions of label encoding or the dimensionality explosion of one-hot encoding.

**Ordered Boosting:** Unlike standard gradient boosting, which uses the same data points to calculate gradients and build trees, CatBoost employs ordered boosting---a permutation-based approach that reduces prediction shift and overfitting. This is particularly valuable for credit risk where training data may contain spurious correlations.

**Symmetric Trees:** CatBoost builds balanced (symmetric) trees by default, which can provide different decision boundaries compared to the asymmetric trees of LightGBM and XGBoost, adding valuable diversity to the ensemble.

**Mathematical Foundation:**

The CatBoost ordered target statistic for a categorical feature is computed as:

$$\hat{x}_{k}^{i} = \frac{\sum_{j=1}^{p-1} [x_{\sigma_{j},k} = x_{\sigma_{p},k}] \cdot y_{\sigma_{j}} + a \cdot P}{\sum_{j=1}^{p-1} [x_{\sigma_{j},k} = x_{\sigma_{p},k}] + a}$$

Where $\sigma$ is a random permutation of training samples, $p$ is the target sample position, $a$ is the prior weight, and $P$ is the prior value.

### LightGBM (Diversity Model)

LightGBM (Microsoft) contributes to ensemble diversity through:
- **Leaf-wise tree growth:** Different tree structure compared to CatBoost's level-wise approach
- **Gradient-based One-Side Sampling (GOSS):** Different sampling strategy for finding best splits
- **Exclusive Feature Bundling:** Efficient handling of sparse features

### XGBoost (Diversity Model)

XGBoost (DMLC) provides:
- **Regularized objective:** Built-in L1/L2 regularization in the objective function
- **Different split finding algorithm:** Exact and approximate greedy algorithms
- **Proven track record:** Extensive validation in credit risk competitions and production systems

### Ensemble Diversity Rationale

The ensemble benefits from **error decorrelation**---different algorithms make different types of errors on different samples. When predictions are averaged, uncorrelated errors tend to cancel out, reducing overall prediction variance:

$$\text{Ensemble Variance} = \frac{1}{M^2} \sum_{i=1}^{M} \text{Var}(f_i) + \frac{2}{M^2} \sum_{i<j} \text{Cov}(f_i, f_j)$$

Where $M$ is the number of models. Lower covariance between models reduces ensemble variance.

## 6.2 Cross-Validation Strategy

**Stratified K-Fold (K=5)** is employed to ensure each fold maintains the same default rate as the full dataset. This is essential for imbalanced credit risk data where random splitting could produce folds with insufficient default samples for reliable evaluation.

**Business Justification:** Credit portfolios exhibit structural stability in default rates. Stratified sampling mimics this stability, ensuring that validation performance estimates reflect production conditions.

## 6.3 Model Configuration

Each algorithm is configured with conservative regularization parameters appropriate for credit risk modeling:

| Parameter | CatBoost | LightGBM | XGBoost | Purpose |
|-----------|----------|----------|---------|---------|
| Learning Rate | 0.02 | 0.02 | 0.02 | Slow, stable convergence |
| Max Depth | 6 | 5 | 5 | Prevent overfitting |
| Regularization | L2: 5.0 | L1: 1.0, L2: 5.0 | L1: 1.0, L2: 5.0 | Penalize complexity |
| Min Samples/Leaf | 50 | 50 | 50 | Require sufficient data per leaf |
| Subsample | 0.7 | 0.7 | 0.7 | Reduce variance |
| Class Weighting | Balanced | scale_pos_weight | scale_pos_weight | Address imbalance |

The conservative regularization reflects the **bias-variance tradeoff** prioritization in credit risk: slightly higher bias (simpler models) is preferred over high variance (unstable predictions) because production models must perform consistently across economic cycles.


```python
# ============================================================
# 4.1 — CATBOOST (PRIMARY MODEL)
# ============================================================
skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)

# CatBoost handles categoricals natively — key advantage!
cat_col_indices = [X_cb.columns.get_loc(c) for c in catboost_cat_cols]

cb_params = {
    'iterations': 3000,
    'learning_rate': 0.02,
    'depth': 6,
    'l2_leaf_reg': 5.0,
    'min_data_in_leaf': 50,
    'random_strength': 0.5,
    'bagging_temperature': 0.3,
    'border_count': 128,
    'random_seed': SEED,
    'verbose': 0,
    'eval_metric': 'AUC',
    'loss_function': 'Logloss',
    'auto_class_weights': 'Balanced',
    'early_stopping_rounds': 200,
    'use_best_model': True,
}

cb_oof = np.zeros(len(X_cb))
cb_test_preds = np.zeros(len(X_cb_test))
cb_models = []
cb_fold_scores = []

print("Training CatBoost with native categoricals...")
print("=" * 60)

for fold, (train_idx, val_idx) in enumerate(skf.split(X_cb, y)):
    X_tr = X_cb.iloc[train_idx]
    y_tr = y.iloc[train_idx]
    X_vl = X_cb.iloc[val_idx]
    y_vl = y.iloc[val_idx]

    train_pool = Pool(X_tr, y_tr, cat_features=catboost_cat_cols)
    val_pool = Pool(X_vl, y_vl, cat_features=catboost_cat_cols)

    model = CatBoostClassifier(**cb_params)
    model.fit(train_pool, eval_set=val_pool, verbose=0)

    val_pred = model.predict_proba(X_vl)[:, 1]
    cb_oof[val_idx] = val_pred
    cb_test_preds += model.predict_proba(X_cb_test)[:, 1] / N_FOLDS

    fold_auc = roc_auc_score(y_vl, val_pred)
    cb_fold_scores.append(fold_auc)
    cb_models.append(model)
    print(f"  Fold {fold+1}: AUC = {fold_auc:.5f} | Best iter: {model.best_iteration_}")

cb_auc = roc_auc_score(y, cb_oof)
print(f"\n🏆 CatBoost OOF AUC: {cb_auc:.5f} (±{np.std(cb_fold_scores):.5f})")
```

# 7. Hyperparameter Optimization

## 7.1 Optuna Framework

Hyperparameter optimization is conducted using Optuna, a state-of-the-art hyperparameter optimization framework that implements Tree-structured Parzen Estimator (TPE) sampling. TPE is a Bayesian optimization algorithm that models the probability of hyperparameter configurations given observed performance, efficiently navigating high-dimensional search spaces.

### Why Hyperparameter Optimization Matters

In credit risk modeling, hyperparameters directly influence:

| Hyperparameter | Credit Risk Impact |
|----------------|-------------------|
| Learning Rate | Convergence stability; too high causes overshooting |
| Tree Depth | Model complexity; too deep leads to overfitting |
| Regularization (L2) | Penalty on extreme coefficients; prevents unstable PD estimates |
| Min Samples per Leaf | Ensures each leaf has statistically meaningful sample size |
| Subsample | Variance reduction; mimics bootstrap uncertainty |

### TPE Algorithm Overview

The Tree-structured Parzen Estimator works by:
1. Dividating observations into "good" (top performers) and "bad" (poor performers)
2. Modeling $p(x|y < y^*)$ and $p(x|y \geq y^*)$ separately
3. Selecting the next hyperparameter configuration that maximizes the expected improvement

This approach is more efficient than grid search or random search, particularly when the number of hyperparameters is large and evaluations are expensive.

### Regularization Philosophy

Credit risk models require **conservative regularization** because:
- **Production stability** is more important than marginal training gains
- **Regulatory scrutiny** demands models that don't overfit to historical patterns
- **Economic cycle robustness** requires models that generalize across different conditions
- **Calibration quality** depends on smooth, well-regularized probability estimates

The regularization parameters are intentionally set to values that may slightly reduce training performance but improve out-of-sample stability---a trade-off that aligns with banking regulatory expectations.


```python
# ============================================================
# 5.1 — OPTUNA HPO FOR CATBOOST (Expanded Search)
# ============================================================
import optuna
from optuna.samplers import TPESampler
optuna.logging.set_verbosity(optuna.logging.WARNING)

def cb_objective(trial):
    params = {
        'iterations': 5000,
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.1, log=True),
        'depth': trial.suggest_int('depth', 3, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 0.1, 30.0, log=True),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 10, 200),
        'random_strength': trial.suggest_float('random_strength', 0.01, 3.0),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0.0, 2.0),
        'border_count': trial.suggest_int('border_count', 32, 255),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.5, 1.0),
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise', 'Lossguide']),
        'random_seed': SEED,
        'verbose': 0,
        'eval_metric': 'AUC',
        'loss_function': 'Logloss',
        'auto_class_weights': 'Balanced',
        'early_stopping_rounds': 200,
        'use_best_model': True,
        'bootstrap_type': 'MVS',
    }
    # Lossguide requires max_leaves
    if params['grow_policy'] == 'Lossguide':
        params['max_leaves'] = trial.suggest_int('max_leaves', 16, 64)

    scores = []
    for train_idx, val_idx in skf.split(X_cb, y):
        X_tr = X_cb.iloc[train_idx]
        y_tr = y.iloc[train_idx]
        X_vl = X_cb.iloc[val_idx]
        y_vl = y.iloc[val_idx]

        train_pool = Pool(X_tr, y_tr, cat_features=catboost_cat_cols)
        val_pool = Pool(X_vl, y_vl, cat_features=catboost_cat_cols)

        model = CatBoostClassifier(**params)
        model.fit(train_pool, eval_set=val_pool, verbose=0)
        pred = model.predict_proba(X_vl)[:, 1]
        scores.append(roc_auc_score(y_vl, pred))

    return np.mean(scores)

print("Running Optuna HPO for CatBoost (10 trials, expanded search)...")
print("This may take 20-40 minutes...")
study = optuna.create_study(direction='maximize', sampler=TPESampler(seed=SEED))
study.optimize(cb_objective, n_trials=1, show_progress_bar=True)

print(f"\nBest trial AUC: {study.best_value:.5f}")
print(f"Best params:")
for k, v in study.best_params.items():
    print(f"  {k}: {v}")
```

```python
# ============================================================
# 5.2 — RETRAIN CATBOOST WITH BEST PARAMS (Multi-Seed Averaging)
# ============================================================
best_cb_params = {
    'iterations': 5000,
    'verbose': 0,
    'eval_metric': 'AUC',
    'loss_function': 'Logloss',
    'auto_class_weights': 'Balanced',
    'early_stopping_rounds': 300,
    'use_best_model': True,
    **study.best_params
}

if best_cb_params.get('grow_policy') != 'Lossguide' and 'max_leaves' in best_cb_params:
    del best_cb_params['max_leaves']

SEEDS = [42]
cb_tuned_oof = np.zeros(len(X_cb))
cb_tuned_test = np.zeros(len(X_cb_test))
cb_tuned_scores = []
cb_tuned_models = []

print(f"Retraining CatBoost with tuned params ({len(SEEDS)}-seed averaging)...")
print("=" * 60)

for seed_i, s in enumerate(SEEDS):
    seed_oof = np.zeros(len(X_cb))
    seed_test = np.zeros(len(X_cb_test))
    seed_scores = []
    best_cb_params['random_seed'] = s

    for fold, (train_idx, val_idx) in enumerate(skf.split(X_cb, y)):
        X_tr = X_cb.iloc[train_idx]
        y_tr = y.iloc[train_idx]
        X_vl = X_cb.iloc[val_idx]
        y_vl = y.iloc[val_idx]

        train_pool = Pool(X_tr, y_tr, cat_features=catboost_cat_cols)
        val_pool = Pool(X_vl, y_vl, cat_features=catboost_cat_cols)

        model = CatBoostClassifier(**best_cb_params)
        model.fit(train_pool, eval_set=val_pool, verbose=0)

        val_pred = model.predict_proba(X_vl)[:, 1]
        seed_oof[val_idx] = val_pred
        seed_test += model.predict_proba(X_cb_test)[:, 1] / N_FOLDS

        fold_auc = roc_auc_score(y_vl, val_pred)
        seed_scores.append(fold_auc)
        if seed_i == 0:
            cb_tuned_models.append(model)

    seed_auc = roc_auc_score(y, seed_oof)
    print(f"  Seed {s}: OOF AUC = {seed_auc:.5f} (folds: {[f'{s:.4f}' for s in seed_scores]})")
    cb_tuned_oof += seed_oof / len(SEEDS)
    cb_tuned_test += seed_test / len(SEEDS)
    cb_tuned_scores.append(seed_auc)

cb_tuned_auc = roc_auc_score(y, cb_tuned_oof)
print(f"\n🏆 Multi-seed CatBoost OOF AUC: {cb_tuned_auc:.5f} (±{np.std(cb_tuned_scores):.5f})")
print(f"Improvement over base CatBoost: {cb_tuned_auc - cb_auc:+.5f}")
```

# 8. Ensemble Construction

## 8.1 Ensemble Methodology

The ensemble strategy employs weighted blending of the three base algorithms, with weights determined through empirical optimization. The rationale for CatBoost-heavy weighting stems from its superior cross-validation performance and unique categorical handling advantages.

### Weight Optimization Strategy

Multiple weight configurations are evaluated:

| Blend Name | CatBoost | LightGBM | XGBoost | Rationale |
|------------|----------|----------|---------|-----------|
| Equal | 0.333 | 0.333 | 0.333 | Maximum diversity, simple |
| CatBoost Heavy | 0.500 | 0.250 | 0.250 | Favors best performer |
| CatBoost Focus | 0.600 | 0.200 | 0.200 | Strong primary emphasis |
| CatBoost Dominant | 0.700 | 0.150 | 0.150 | Near-single-model with diversity |
| AUC Weighted | Proportional | Proportional | Proportional | Performance-based |

### Mathematical Formulation

The ensemble prediction is computed as:

$$\hat{y}_{ensemble} = w_{cb} \cdot \hat{y}_{cb} + w_{lgb} \cdot \hat{y}_{lgb} + w_{xgb} \cdot \hat{y}_{xgb}$$

Subject to:

$$w_{cb} + w_{lgb} + w_{xgb} = 1, \quad w_i \geq 0$$

The optimal weights are selected to maximize OOF AUC.

## 8.2 Pseudo-Labeling

Pseudo-labeling is a semi-supervised learning technique where high-confidence predictions on unlabeled test data are incorporated into the training set. This approach can improve model performance when the test distribution contains patterns not fully represented in the training data.

### Methodology

1. Generate predictions on the test set using the trained ensemble
2. Select samples with high confidence: predicted probability <= 0.10 (confident non-defaults) or >= 0.60 (confident defaults)
3. Assign pseudo-labels based on the confidence threshold
4. Retrain the model on the combined dataset (original training + pseudo-labeled test)
5. Evaluate whether pseudo-labeled training improves validation performance

### Risk Considerations

Pseudo-labeling introduces potential risks:
- **Confirmation bias:** If initial predictions are wrong, pseudo-labeling amplifies errors
- **Distribution drift:** Pseudo-labels may reinforce existing biases
- **Information leakage:** Without proper validation framing, leakage can occur

The implementation mitigates these risks by:
- Using conservative confidence thresholds (0.10 and 0.60)
- Only adopting pseudo-label predictions if they demonstrably improve OOF AUC
- Maintaining the original validation fold structure for unbiased evaluation

**Banking Context:** In production, semi-supervised techniques require careful governance approval, as they effectively change the training data distribution. For regulatory models, any pseudo-labeling approach must be documented, validated, and subject to ongoing monitoring.


```python
# ============================================================
# 6.1 — BLENDING EXPERIMENTS
# ============================================================
print("Ensemble Blending Experiments")
print("=" * 60)

# Try different weight combinations
blends = {
    'Equal (1:1:1)': (1/3, 1/3, 1/3),
    'CatBoost Heavy (0.5:0.25:0.25)': (0.5, 0.25, 0.25),
    'CatBoost Focus (0.6:0.2:0.2)': (0.6, 0.2, 0.2),
    'CatBoost Dominant (0.7:0.15:0.15)': (0.7, 0.15, 0.15),
    'CatBoost Only (1:0:0)': (1.0, 0.0, 0.0),
    'AUC Weighted': None,  # Will compute
}

# AUC weighted
aucs = np.array([cb_tuned_auc, lgb_auc, xgb_auc])
auc_weights = aucs / aucs.sum()
blends['AUC Weighted'] = tuple(auc_weights)

best_blend_name = None
best_blend_auc = 0

for name, (w_cb, w_lgb, w_xgb) in blends.items():
    blend_oof = w_cb * cb_tuned_oof + w_lgb * lgb_oof + w_xgb * xgb_oof
    blend_auc = roc_auc_score(y, blend_oof)
    marker = "  ★ BEST" if blend_auc > best_blend_auc else ""
    print(f"  {name:35s} | AUC = {blend_auc:.5f}{marker}")
    if blend_auc > best_blend_auc:
        best_blend_auc = blend_auc
        best_blend_name = name
        best_weights = (w_cb, w_lgb, w_xgb)

print(f"\n🏆 Best blend: {best_blend_name} (AUC = {best_blend_auc:.5f})")
print(f"   Weights: CatBoost={best_weights[0]:.2f}, LGB={best_weights[1]:.2f}, XGB={best_weights[2]:.2f}")

# Generate final predictions with best blend
w_cb, w_lgb, w_xgb = best_weights
final_oof = w_cb * cb_tuned_oof + w_lgb * lgb_oof + w_xgb * xgb_oof
final_test = w_cb * cb_tuned_test + w_lgb * lgb_test_preds + w_xgb * xgb_test_preds

final_auc = roc_auc_score(y, final_oof)
print(f"\nFinal Ensemble OOF AUC: {final_auc:.5f}")
```

```python
# ============================================================
# 6.5 — PSEUDO-LABELING
# ============================================================
# Use final ensemble to assign pseudo-labels to confident test predictions
confidence_threshold_low = 0.10   # confident non-defaults
confidence_threshold_high = 0.60  # confident defaults

pseudo_default = final_test >= confidence_threshold_high
pseudo_nondefault = final_test <= confidence_threshold_low
pseudo_mask = pseudo_default | pseudo_nondefault

print(f"Pseudo-labeling statistics:")
print(f"  Confident defaults (>={confidence_threshold_high}): {pseudo_default.sum():,}")
print(f"  Confident non-defaults (<={confidence_threshold_low}): {pseudo_nondefault.sum():,}")
print(f"  Total pseudo-labeled: {pseudo_mask.sum():,} / {len(final_test):,} ({pseudo_mask.mean():.1%})")

if pseudo_mask.sum() > 100:
    # Build pseudo-labeled dataset
    X_pseudo = X_cb_test[pseudo_mask].copy()
    y_pseudo = (final_test[pseudo_mask] >= confidence_threshold_high).astype(int)

    print(f"  Pseudo-label default rate: {y_pseudo.mean():.2%}")

    # Combine train + pseudo-labeled
    X_combined = pd.concat([X_cb, X_pseudo], axis=0, ignore_index=True)
    y_pseudo = pd.Series(
        (final_test[pseudo_mask] >= confidence_threshold_high).astype(int),
        name=TARGET
    )

    y_combined = pd.concat([y, y_pseudo], axis=0, ignore_index=True)

    # Retrain CatBoost on combined data with same best params
    pl_oof = np.zeros(len(X_cb))
    pl_test = np.zeros(len(X_cb_test))
    pl_scores = []

    best_cb_params['random_seed'] = SEED

    print("\nRetraining CatBoost with pseudo-labels...")
    for fold, (train_idx, val_idx) in enumerate(skf.split(X_cb, y)):
        # Use original val fold for evaluation, but augment train with pseudo-labels
        X_tr_orig = X_cb.iloc[train_idx]
        y_tr_orig = y.iloc[train_idx]
        X_vl = X_cb.iloc[val_idx]
        y_vl = y.iloc[val_idx]

        # Augment training with pseudo-labels
        X_tr_aug = pd.concat([X_tr_orig, X_pseudo], axis=0, ignore_index=True)
        y_tr_aug = pd.concat([y_tr_orig, pd.Series(y_pseudo.values)], axis=0, ignore_index=True)

        train_pool = Pool(X_tr_aug, y_tr_aug, cat_features=catboost_cat_cols)
        val_pool = Pool(X_vl, y_vl, cat_features=catboost_cat_cols)

        model = CatBoostClassifier(**best_cb_params)
        model.fit(train_pool, eval_set=val_pool, verbose=0)

        val_pred = model.predict_proba(X_vl)[:, 1]
        pl_oof[val_idx] = val_pred
        pl_test += model.predict_proba(X_cb_test)[:, 1] / N_FOLDS

        fold_auc = roc_auc_score(y_vl, val_pred)
        pl_scores.append(fold_auc)
        print(f"  Fold {fold+1}: AUC = {fold_auc:.5f}")

    pl_auc = roc_auc_score(y, pl_oof)
    print(f"\n🏆 Pseudo-label CatBoost OOF AUC: {pl_auc:.5f}")

    # Only use pseudo-label predictions if they improve the score
    if pl_auc > cb_tuned_auc:
        print(f"✅ Pseudo-labeling improved AUC by {pl_auc - cb_tuned_auc:+.5f}")
        print("   Using pseudo-label predictions in final blend.")
        # Blend pseudo-label preds into the final test predictions
        final_test = 0.5 * final_test + 0.5 * pl_test
        final_oof_pl = w_cb * pl_oof + w_lgb * lgb_oof + w_xgb * xgb_oof
        if roc_auc_score(y, final_oof_pl) > final_auc:
            final_oof = final_oof_pl
            final_auc = roc_auc_score(y, final_oof)
            print(f"   Updated final OOF AUC: {final_auc:.5f}")
    else:
        print(f"⚠️ Pseudo-labeling did NOT improve (delta: {pl_auc - cb_tuned_auc:+.5f}). Keeping original.")
else:
    print("Not enough confident predictions for pseudo-labeling. Skipping.")
```

# 9. Comprehensive Model Evaluation

## 9.1 Evaluation Framework

Model evaluation in credit risk requires a multi-dimensional assessment that goes beyond a single metric. This section implements comprehensive diagnostics examining discriminatory power, calibration quality, probability distribution characteristics, and classification performance at optimized thresholds.

### Evaluation Metrics and Their Banking Significance

| Metric | Formula | Banking Interpretation |
|--------|---------|----------------------|
| **ROC-AUC** | $\int_0^1 \text{TPR}(t) \, d\text{FPR}(t)$ | Overall discriminatory power; probability that a random defaulter is scored higher than a random non-defaulter |
| **PR-AUC** | $\int_0^1 \text{Precision}(r) \, dr$ | Discriminatory power accounting for class imbalance; more informative than ROC-AUC for rare defaults |
| **Log Loss** | $-\frac{1}{N}\sum [y \log(\hat{p}) + (1-y)\log(1-\hat{p})]$ | Penalty for confident wrong predictions; sensitive to probability calibration |
| **Brier Score** | $\frac{1}{N}\sum (y - \hat{p})^2$ | Mean squared error of probability predictions; measures calibration quality |
| **KS Statistic** | $\max_t |\text{TPR}(t) - \text{FPR}(t)|$ | Maximum separation between cumulative distributions of scores for defaulters and non-defaulters |
| **Gini Coefficient** | $2 \times \text{AUC} - 1$ | Linear transformation of AUC commonly used in banking; ranges from 0 (no discrimination) to 1 (perfect) |

### The Gini Coefficient in Banking

The Gini coefficient is the preferred metric in many banking contexts because:
- It has an intuitive scale (0% to 100%)
- It directly relates to the Lorenz curve interpretation
- Regulatory benchmarks are often expressed in Gini terms
- A Gini above 40% is generally considered acceptable for retail credit models
- A Gini above 60% is considered excellent

$$\text{Gini} = 2 \times \text{AUC} - 1$$

### Threshold Optimization

The optimal classification threshold is determined by maximizing the F1-Score, which balances precision and recall. However, in practice, the threshold should be adjusted based on the **relative costs** of false positives and false negatives:

$$\text{Optimal Threshold} = \arg\max_t \left( \frac{2 \cdot \text{Precision}(t) \cdot \text{Recall}(t)}{\text{Precision}(t) + \text{Recall}(t)} \right)$$

**Banking Considerations for Threshold Selection:**

| Threshold Strategy | Use Case | Risk Appetite |
|-------------------|----------|---------------|
| Low Threshold (0.15-0.20) | Maximum default capture; conservative lending | Very low risk tolerance |
| Medium Threshold (0.25-0.35) | Balanced approach; standard underwriting | Moderate risk tolerance |
| High Threshold (0.40-0.50) | Minimize false positives; maximize approval rate | Higher risk tolerance |

The cost asymmetry in lending (false negatives typically cost 3-5x more than false positives) generally favors lower thresholds that maximize recall.

## 9.2 Calibration Assessment

Probability calibration ensures that the model's predicted default probabilities match the observed default frequencies. For example, among loans with a predicted PD of 20%, approximately 20% should actually default.

### Calibration Methods

The model incorporates calibration through:
1. **Platt Scaling:** Logistic regression on model outputs
2. **Isotonic Regression:** Non-parametric calibration mapping
3. **Built-in Calibration:** CatBoost's probability estimates are naturally well-calibrated due to the ordered boosting mechanism

### Reliability Diagrams

A reliability diagram plots predicted probabilities against observed frequencies. A perfectly calibrated model follows the diagonal line. Deviations indicate:
- **Above diagonal:** Model is underconfident (predictions too low)
- **Below diagonal:** Model is overconfident (predictions too high)

Well-calibrated probabilities are essential for:
- IFRS 9 expected credit loss calculations
- Risk-based pricing decisions
- Portfolio loss forecasting
- Regulatory reporting

# 10. SHAP Explainability Analysis

## 10.1 Introduction to SHAP

SHAP (SHapley Additive exPlanations) is a unified framework for interpreting machine learning model predictions based on cooperative game theory. Developed by Lundberg and Lee (2017), SHAP provides both global (model-level) and local (individual prediction-level) interpretability.

### Mathematical Foundation

SHAP values are derived from Shapley values in cooperative game theory. For a prediction $f(x)$, the SHAP value for feature $i$ is:

$$\phi_i = \sum_{S \subseteq N \setminus \{i\}} \frac{|S|!(|N| - |S| - 1)!}{|N|!} [f_{S \cup \{i\}}(x_{S \cup \{i\}}) - f_S(x_S)]$$

Where:
- $N$ is the set of all features
- $S$ is a subset of features not containing feature $i$
- $f_S(x_S)$ is the model prediction using only features in subset $S$

This formula computes the marginal contribution of feature $i$ across all possible feature coalitions, weighted by the coalition size.

### Key Properties

SHAP values satisfy four desirable properties:

1. **Efficiency:** $\sum_i \phi_i = f(x) - E[f(x)]$ --- the SHAP values sum to the prediction minus the baseline
2. **Symmetry:** If two features contribute equally, they receive equal SHAP values
3. **Dummy:** A feature that doesn't change predictions receives a SHAP value of zero
4. **Additivity:** For a model that is a sum of models, SHAP values are the sum of the component SHAP values

### Why SHAP Matters for Banking

| Regulatory Requirement | SHAP Capability |
|----------------------|----------------|
| Model transparency | Global feature importance ranking |
| Individual decision explanation | Local attribution per loan application |
| Fairness assessment | Identifies biased feature contributions |
| Model validation | Validates that important features are economically intuitive |
| Audit trail | Documented explanation for each prediction |

The European Banking Authority's Guidelines on PD and LGD estimation (EBA/GL/2017/16) implicitly require that credit risk models be "transparent and comprehensible." SHAP analysis directly addresses this requirement by providing:

- **Feature Importance Ranking:** Which variables drive predictions?
- **Direction of Effect:** How does each feature influence the default probability?
- **Interaction Effects:** Do features work together in non-additive ways?
- **Individual Explanations:** Why was this specific loan assigned this probability?

### Global vs. Local Interpretability

**Global Interpretability (Summary Plot):**
- Shows the distribution of SHAP values for each feature across the dataset
- Reveals which features are most important overall
- Indicates the direction of effect (positive or negative contribution to default risk)
- Identifies non-linear relationships and interaction effects

**Local Interpretability (Force/Waterfall Plot):**
- Explains a single prediction by showing how each feature pushes the prediction
- Enables personalized explanations for declined loan applications
- Supports fair lending compliance by documenting decision rationale
- Facilitates appeals and override processes


```python
# ============================================================
# 8.1 — SHAP VALUES
# ============================================================
try:
    import shap
    print("Computing SHAP values...")

    sample_size = min(2000, len(X_cb))
    sample_idx = np.random.choice(len(X_cb), size=sample_size, replace=False)
    X_shap = X_cb.iloc[sample_idx]

    explainer = shap.TreeExplainer(best_cb_model)
    shap_values = explainer.shap_values(X_shap)

    # Summary plot
    plt.figure(figsize=(12, 10))
    shap.summary_plot(shap_values, X_shap, max_display=20, show=False)
    plt.title('SHAP Feature Impact', fontweight='bold')
    plt.tight_layout()
    plt.show()

    print("SHAP analysis complete.")
except Exception as e:
    print(f"SHAP analysis skipped: {e}")
```

# 11. Statistical Testing and Model Comparison

## 11.1 Friedman Test

The Friedman test is a non-parametric statistical test used to compare the performance of multiple models across multiple datasets (or folds). Unlike repeated measures ANOVA, the Friedman test does not assume normality of the differences, making it appropriate for cross-validation performance metrics.

### Test Statistic

The Friedman test statistic is computed as:

$$\chi_F^2 = \frac{12N}{k(k+1)} \left[ \sum_{j=1}^{k} R_j^2 - \frac{k(k+1)^2}{4} \right]$$

Where:
- $N$ is the number of folds (5)
- $k$ is the number of models (3)
- $R_j$ is the average rank of model $j$ across all folds

The null hypothesis states that all models perform equivalently (their ranks are not systematically different). Rejection of the null hypothesis indicates that at least one model performs significantly differently from the others.

**Banking Interpretation:** Statistical significance testing provides evidence that model selection is based on genuine performance differences rather than random variation. This is critical for model governance documentation and regulatory submissions.

## 11.2 Nemenyi Post-Hoc Test

When the Friedman test indicates significant differences, the Nemenyi post-hoc test identifies which specific pairs of models differ significantly.

### Critical Difference (CD)

The Critical Difference is computed as:

$$\text{CD} = q_\alpha \sqrt{\frac{k(k+1)}{6N}}$$

Where $q_\alpha$ is the critical value from the studentized range distribution at significance level $\alpha$ (typically 0.05).

Two models are significantly different if their average rank difference exceeds the Critical Difference.

### Critical Difference Diagram

A Critical Difference diagram provides a visual summary of the statistical comparison:
- Models are plotted on a line ordered by average rank
- Models connected by a thick horizontal bar are not significantly different
- Models not connected are statistically different

**Interpretation for Banking:** The CD diagram quickly communicates whether the chosen primary model (CatBoost) is statistically superior to alternatives, supporting the model selection rationale in governance documentation.

## 11.3 Cross-Validation Comparison Summary

The cross-validation scores provide the raw data for statistical testing. Each model's fold scores represent independent performance estimates on held-out data.

| Model | Fold 1 | Fold 2 | Fold 3 | Fold 4 | Fold 5 | Mean | Std |
|-------|--------|--------|--------|--------|--------|------|-----|
| CatBoost (Tuned) | ... | ... | ... | ... | ... | ... | ... |
| LightGBM | ... | ... | ... | ... | ... | ... | ... |
| XGBoost | ... | ... | ... | ... | ... | ... | ... |

The Friedman test applied to these fold scores determines whether the observed performance differences are statistically significant or attributable to random variation.


```python
# ============================================================
# STATISTICAL TESTING: FRIEDMAN TEST & NEMENYI POST-HOC
# ============================================================
from scipy.stats import friedmanchisquare
from scipy.stats import rankdata
import itertools

try:
    # Prepare fold scores for all models
    fold_scores_df = pd.DataFrame({
        'Fold': range(1, N_FOLDS + 1),
        'CatBoost_Tuned': cb_tuned_scores if len(cb_tuned_scores) == N_FOLDS else [cb_tuned_auc]*N_FOLDS,
        'LightGBM': lgb_fold_scores,
        'XGBoost': xgb_fold_scores
    })

    # Use per-fold scores for CatBoost (from first seed) if available
    if len(cb_fold_scores) == N_FOLDS:
        fold_scores_df['CatBoost_Base'] = cb_fold_scores

    print("Cross-Validation Fold Scores:")
    print(fold_scores_df.to_string(index=False))

    # Friedman Test on base models
    stat, p_value = friedmanchisquare(lgb_fold_scores, xgb_fold_scores, cb_fold_scores)
    print(f"\n{'='*60}")
    print("FRIEDMAN TEST RESULTS")
    print(f"{'='*60}")
    print(f"Test Statistic: {stat:.4f}")
    print(f"P-Value: {p_value:.6f}")
    print(f"Significance Level: α = 0.05")
    if p_value < 0.05:
        print("Result: REJECT null hypothesis — models differ significantly")
    else:
        print("Result: FAIL TO REJECT null hypothesis — no significant difference")

    # Compute ranks per fold
    ranks = np.zeros((N_FOLDS, 3))
    all_scores = np.column_stack([lgb_fold_scores, xgb_fold_scores, cb_fold_scores])
    for i in range(N_FOLDS):
        ranks[i] = rankdata(-all_scores[i])  # Negative for higher-is-better

    avg_ranks = ranks.mean(axis=0)
    model_names = ['LightGBM', 'XGBoost', 'CatBoost']

    print(f"\n{'='*60}")
    print("AVERAGE RANKS (lower is better)")
    print(f"{'='*60}")
    for name, rank in zip(model_names, avg_ranks):
        print(f"  {name:15s}: {rank:.2f}")

    # Nemenyi Critical Difference
    k = 3  # number of models
    N_folds = N_FOLDS
    q_alpha_05 = 2.343  # q-value for k=3, alpha=0.05
    cd = q_alpha_05 * np.sqrt(k * (k + 1) / (6 * N_folds))

    print(f"\n{'='*60}")
    print("NEMENYI POST-HOC TEST")
    print(f"{'='*60}")
    print(f"Critical Difference (α=0.05): {cd:.4f}")
    print(f"\nPairwise Comparisons:")

    for (i, mi), (j, mj) in itertools.combinations(enumerate(model_names), 2):
        diff = abs(avg_ranks[i] - avg_ranks[j])
        sig = "*** SIGNIFICANT" if diff > cd else "Not significant"
        print(f"  {mi} vs {mj}: |{avg_ranks[i]:.2f} - {avg_ranks[j]:.2f}| = {diff:.2f}  {sig}")

except Exception as e:
    print(f"Statistical testing section: {e}")
    print("Note: Friedman test requires per-fold scores from the models.")
```

# 12. Error Analysis and Risk Interpretation

## 12.1 Classification of Prediction Errors

In credit risk modeling, understanding prediction errors is as important as understanding correct predictions. Errors fall into two categories with fundamentally different business implications:

### False Negatives (Type II Errors): Missed Defaults

A false negative occurs when the model predicts a low default probability for a loan that ultimately defaults. These are the most dangerous and costly errors in credit risk.

**Business Impact:**
- Direct loss of principal and accrued interest
- Collection and legal expenses
- Opportunity cost of capital tied up in non-performing loans
- Potential increase in regulatory provisions
- Reputational risk if default rate exceeds expectations

**Financial Cost Model:**

$$\text{Cost}_{FN} = \text{LGD} \times \text{EAD} + \text{Collection Cost} + \text{Legal Cost}$$

For a typical consumer loan, the total cost of a false negative can range from 30% to 80% of the exposure amount.

### False Positives (Type I Errors): Incorrect Declines

A false positive occurs when the model predicts a high default probability for a loan that would have performed well.

**Business Impact:**
- Lost interest income over the loan term
- Damaged customer relationship and potential lifetime value loss
- Competitive disadvantage if approval rates are too low
- Regulatory scrutiny if decline patterns suggest discriminatory practices

**Financial Cost Model:**

$$\text{Cost}_{FP} = \text{Interest Income}_{lost} + \text{Relationship Value}_{lost}$$

The cost of a false positive is typically lower than a false negative but can be substantial for high-value, long-term relationships.

### Asymmetric Cost Ratio

In most lending contexts, the cost ratio is:

$$\frac{\text{Cost}_{FN}}{\text{Cost}_{FP}} \approx 3 	ext{ to } 5$$

This asymmetry explains why credit risk models typically optimize for **recall** (sensitivity) rather than **precision**, accepting more false positives to minimize missed defaults.

## 12.2 Error Analysis by Segment

Understanding where the model makes errors is critical for:
- **Targeted model improvement:** Focus refinement on high-error segments
- **Policy development:** Apply additional scrutiny to error-prone categories
- **Portfolio monitoring:** Identify segments requiring enhanced watchlist monitoring

### Key Segmentation Dimensions

| Dimension | Analysis | Action |
|-----------|----------|--------|
| DTI Ratio | High errors at extreme DTI values | Enhanced validation for edge cases |
| Loan Purpose | Certain purposes may be harder to predict | Specialized sub-models |
| Province | Regional economic differences | Geographic risk overlays |
| Employment Sector | Informal sector prediction difficulty | Alternative data integration |
| Loan Amount | Very small or very large loans | Amount-band specific thresholds |

## 12.3 Uncertainty Analysis

Predictions near the decision threshold (typically 0.25-0.40) represent the highest uncertainty. These "gray zone" cases benefit most from:
- Manual underwriter review
- Additional documentation requirements
- Conditional approvals with covenant structures
- Higher pricing to compensate for uncertainty

### Prediction Distribution Analysis

The distribution of predicted probabilities reveals model confidence:
- **Bimodal distribution:** Model is confident (good calibration)
- **Uniform distribution:** Model is uncertain (may need more features)
- **Concentrated near threshold:** Many cases require manual review

# 13. IFRS 9 Staging and Expected Credit Loss Framework

## 13.1 IFRS 9 Staging Methodology

IFRS 9 requires loans to be classified into one of three stages based on their credit risk characteristics. The staging determination directly drives the measurement approach for expected credit losses.

### Stage Definitions

| Stage | Name | Criteria | ECL Horizon | Interest Recognition |
|-------|------|----------|-------------|-------------------|
| Stage 1 | Performing | No significant increase in credit risk since origination | 12-month | Gross carrying amount |
| Stage 2 | Underperforming | Significant increase in credit risk (SICR) observed | Lifetime | Gross carrying amount |
| Stage 3 | Non-performing | Objective evidence of impairment/credit impairment | Lifetime | Net carrying amount |

### SICR Assessment Using Model Scores

The model's predicted probabilities of default provide a quantitative foundation for SICR assessment. The approach compares the origination PD with the current PD:

$$\text{SICR Triggered if:} \quad \text{Current PD} > \text{Origination PD} + \text{Threshold}$$

Common SICR thresholds in banking practice:
- Absolute PD increase: > 50 basis points (0.50%)
- Relative PD increase: > 2x the origination PD
- Migration to a higher risk grade

### Model-Based Staging Assignment

For this portfolio, we implement a conceptual staging framework using the predicted probabilities:

| Stage | PD Range | Interpretation | ECL Calculation |
|-------|----------|----------------|-----------------|
| Stage 1 | PD < 15% | Low to moderate risk | 12-month ECL |
| Stage 2 | 15% <= PD < 50% | Elevated risk | Lifetime ECL |
| Stage 3 | PD >= 50% | High impairment risk | Lifetime ECL (credit-impaired) |

## 13.2 Expected Credit Loss Calculation

### The ECL Formula

$$\text{ECL} = \sum_{t=1}^{T} \text{PD}_t \times \text{LGD} \times \text{EAD}_t \times \text{DF}_t$$

Where:
- $\text{PD}_t$: Marginal probability of default in period $t$
- $\text{LGD}$: Loss given default (assumed constant or varying)
- $\text{EAD}_t$: Exposure at default in period $t$
- $\text{DF}_t$: Discount factor for period $t$
- $T$: Time horizon (12 months for Stage 1, lifetime for Stage 2/3)

### Simplified Demonstration

For practical demonstration with the available data, we use a simplified single-period ECL:

$$\text{ECL} \approx \text{PD} \times \text{LGD} \times \text{EAD}$$

Where:
- PD = Model predicted probability
- LGD = Assumed 40% (typical for unsecured consumer lending)
- EAD = Current loan balance (amount_usd)

### Portfolio-Level ECL Aggregation

$$\text{Total ECL} = \sum_{i=1}^{N} \text{ECL}_i = \sum_{i=1}^{N} \text{PD}_i \times \text{LGD}_i \times \text{EAD}_i$$

## 13.3 Stage Distribution and Migration Analysis

The stage distribution provides a snapshot of portfolio health and drives the overall provision level.

### Expected Stage Distribution (Typical Retail Portfolio)

| Stage | Typical Range | Interpretation |
|-------|--------------|----------------|
| Stage 1 | 70-85% | Core performing portfolio |
| Stage 2 | 10-25% | Watchlist, enhanced monitoring |
| Stage 3 | 3-8% | Impaired, collection action |

A portfolio with >90% Stage 1 loans is considered healthy, while Stage 3 above 10% indicates significant credit quality deterioration.

## 13.4 Business Implications of Staging

### Stage Migration Triggers Business Actions

| Migration | Business Response |
|-----------|-------------------|
| Stage 1 -> Stage 2 | Enhanced monitoring, potential covenant review, watchlist placement |
| Stage 2 -> Stage 3 | Impairment recognition, collections engagement, restructuring evaluation |
| Stage 2 -> Stage 1 | SICR reversal (if credit improvement is sustained), reduced monitoring |

### Provision Impact

The staging distribution directly affects the provision-to-loan ratio:

$$\text{Provision Ratio} = \frac{\sum \text{ECL}}{\sum \text{EAD}}$$

Higher Stage 2 and Stage 3 concentrations increase the provision ratio, reducing reported earnings and regulatory capital ratios.


```python
# ============================================================
# IFRS 9 STAGING & ECL SIMULATION
# ============================================================

# Assumptions for demonstration
LGD_ASSUMPTION = 0.40  # 40% loss given default (typical unsecured consumer)
DISCOUNT_RATE = 0.15   # 15% annual discount rate (reflecting high interest rate environment)

# Create staging dataset
staging_df = pd.DataFrame({
    'ID': train_df[ID_COL].values,
    'amount_usd': train_df['amount_usd'].values,
    'pd_12m': final_oof,  # Using model output as 12-month PD
    'actual_default': y.values
})

# Assign IFRS 9 stages
# Stage 1: PD < 15% (performing)
# Stage 2: 15% <= PD < 50% (SICR)
# Stage 3: PD >= 50% (credit impaired)
def assign_stage(pd):
    if pd < 0.15:
        return 'Stage 1: Performing'
    elif pd < 0.50:
        return 'Stage 2: SICR'
    else:
        return 'Stage 3: Impaired'

staging_df['stage'] = staging_df['pd_12m'].apply(assign_stage)

# Calculate ECL
staging_df['lgd'] = LGD_ASSUMPTION
staging_df['ead'] = staging_df['amount_usd']
staging_df['ecl'] = staging_df['pd_12m'] * staging_df['lgd'] * staging_df['ead']

# Stage Distribution Analysis
print("=" * 60)
print("IFRS 9 STAGE DISTRIBUTION ANALYSIS")
print("=" * 60)

stage_summary = staging_df.groupby('stage').agg({
    'ID': 'count',
    'amount_usd': 'sum',
    'ecl': 'sum',
    'pd_12m': 'mean',
    'actual_default': 'mean'
}).rename(columns={'ID': 'Loan_Count', 'amount_usd': 'Total_EAD', 
                   'ecl': 'Total_ECL', 'pd_12m': 'Avg_PD', 'actual_default': 'Actual_Default_Rate'})

stage_summary['EAD_Pct'] = stage_summary['Total_EAD'] / stage_summary['Total_EAD'].sum()
stage_summary['Provision_Rate'] = stage_summary['Total_ECL'] / stage_summary['Total_EAD']
stage_summary['Loan_Count_Pct'] = stage_summary['Loan_Count'] / stage_summary['Loan_Count'].sum()

print(stage_summary.to_string())

# Portfolio totals
total_ead = staging_df['ead'].sum()
total_ecl = staging_df['ecl'].sum()
portfolio_provision_rate = total_ecl / total_ead

print(f"\n{'='*60}")
print("PORTFOLIO SUMMARY")
print(f"{'='*60}")
print(f"Total Exposure (EAD): ${total_ead:,.2f}")
print(f"Total Expected Credit Loss (ECL): ${total_ecl:,.2f}")
print(f"Portfolio Provision Rate: {portfolio_provision_rate:.2%}")
print(f"Weighted Average PD: {staging_df['pd_12m'].mean():.2%}")
print(f"Actual Default Rate: {y.mean():.2%}")
print(f"ECL Calibration: Predicted/Actual = {staging_df['pd_12m'].mean() / y.mean():.2f}x")

# Stage distribution visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Stage distribution by loan count
stage_counts = staging_df['stage'].value_counts()
colors_stage = [COLORS['success'], '#F39C12', COLORS['danger']]
axes[0,0].pie(stage_counts, labels=stage_counts.index, autopct='%1.1f%%', 
              colors=colors_stage, startangle=90)
axes[0,0].set_title('Stage Distribution by Loan Count', fontweight='bold')

# Stage distribution by EAD
stage_ead = staging_df.groupby('stage')['ead'].sum()
axes[0,1].pie(stage_ead, labels=stage_ead.index, autopct='%1.1f%%', 
              colors=colors_stage, startangle=90)
axes[0,1].set_title('Stage Distribution by Exposure (EAD)', fontweight='bold')

# ECL by stage
stage_ecl = staging_df.groupby('stage')['ecl'].sum()
axes[1,0].bar(range(len(stage_ecl)), stage_ecl.values, color=colors_stage, edgecolor='white')
axes[1,0].set_xticks(range(len(stage_ecl)))
axes[1,0].set_xticklabels(stage_ecl.index, rotation=15, ha='right')
axes[1,0].set_ylabel('Expected Credit Loss ($)')
axes[1,0].set_title('ECL by IFRS 9 Stage', fontweight='bold')

# Provision rate by stage
prov_rates = stage_summary['Provision_Rate']
axes[1,1].bar(range(len(prov_rates)), prov_rates.values, color=colors_stage, edgecolor='white')
axes[1,1].set_xticks(range(len(prov_rates)))
axes[1,1].set_xticklabels(prov_rates.index, rotation=15, ha='right')
axes[1,1].set_ylabel('Provision Rate (ECL/EAD)')
axes[1,1].set_title('Provision Rate by Stage', fontweight='bold')
axes[1,1].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

plt.suptitle('IFRS 9 Staging & Expected Credit Loss Analysis', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Risk Heatmap: Province x Product
print("\n" + "=" * 60)
print("RISK HEATMAP: PROVINCE x PRODUCT (Avg PD)")
print("=" * 60)
heatmap_data = train_df.copy()
heatmap_data['pd'] = final_oof
heatmap_data['product_code'] = heatmap_data['product_code'].astype(str)
heatmap_pivot = heatmap_data.groupby(['province', 'product_code'])['pd'].mean().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(heatmap_pivot, annot=True, fmt='.2%', cmap='YlOrRd', ax=ax, cbar_kws={'label': 'Avg PD'})
ax.set_title('Risk Heatmap: Average PD by Province and Product', fontweight='bold')
ax.set_xlabel('Product Code')
ax.set_ylabel('Province')
plt.tight_layout()
plt.show()
```

# 14. Model Governance and Monitoring Framework

## 14.1 Model Governance Overview

Model governance in banking refers to the comprehensive framework of policies, procedures, and controls that ensure credit risk models are developed, validated, and used appropriately. The three lines of defense model is the industry standard:

| Line of Defense | Role | Responsibility |
|----------------|------|---------------|
| 1st Line: Model Development | Data Scientists, Quants | Model design, development, testing |
| 2nd Line: Model Validation | Independent Validation Team | Independent review, challenge, approval |
| 3rd Line: Internal Audit | Audit Department | Periodic review of governance framework |

### Regulatory Framework

Model risk management in banking is governed by:
- **SR 11-7:** Federal Reserve guidance on model risk management
- **EBA Guidelines:** European Banking Authority guidelines on internal governance
- **BCBS 239:** Principles for effective risk data aggregation
- **IFRS 9:** Financial reporting standard requiring transparent, defensible models

## 14.2 Model Risk Rating

Models are classified by risk tier based on materiality and complexity:

| Risk Tier | Description | Validation Frequency |
|-----------|-------------|---------------------|
| High | Material P&L impact, complex methodology | Annual full validation |
| Medium | Moderate impact, standard techniques | Biennial full validation |
| Low | Immaterial impact, simple approaches | Triennial validation |

A credit risk PD model used for both IFRS 9 provisioning and regulatory capital would typically be classified as **High Risk Tier**.

## 14.3 Model Monitoring Framework

Production models require ongoing monitoring to detect performance degradation. The monitoring framework includes:

### Performance Monitoring

| Metric | Frequency | Action Trigger |
|--------|-----------|----------------|
| ROC-AUC | Monthly | Decline > 5% from baseline |
| KS Statistic | Monthly | Decline > 10% from baseline |
| Gini Coefficient | Monthly | Decline > 5% from baseline |
| Calibration Error | Quarterly | Brier score increase > 20% |

### Stability Monitoring

| Metric | Frequency | Action Trigger |
|--------|-----------|----------------|
| Population Stability Index (PSI) | Monthly | PSI > 0.25 (significant shift) |
| Characteristic Stability Index (CSI) | Monthly | Any CSI > 0.10 |
| Score Distribution | Monthly | > 10% shift in score bands |

### Population Stability Index (PSI)

PSI measures the distributional shift between a baseline (training) population and a current (production) population:

$$\text{PSI} = \sum_{i=1}^{n} (P_i - Q_i) \times \ln\left(\frac{P_i}{Q_i}\right)$$

Where $P_i$ is the percentage of observations in bin $i$ for the baseline, and $Q_i$ is the percentage for the current population.

| PSI Value | Interpretation | Action |
|-----------|---------------|--------|
| < 0.10 | Negligible shift | Continue monitoring |
| 0.10 - 0.25 | Moderate shift | Investigate, prepare contingency |
| > 0.25 | Significant shift | Urgent model review required |

## 14.4 Challenger vs. Champion Framework

A robust model governance framework maintains a "champion" (production) model and one or more "challenger" models:

| Role | Purpose | Trigger for Promotion |
|------|---------|----------------------|
| Champion | Production predictions | Baseline for comparison |
| Challenger A | Alternative algorithm | Outperforms champion for 3+ months |
| Challenger B | Alternative features | Outperforms champion for 3+ months |

The challenger framework ensures continuous improvement and provides fallback options if the champion model degrades.

## 14.5 Recalibration Policy

Models should be recalibrated when:
1. PSI exceeds 0.25 for two consecutive months
2. AUC declines by more than 5% from validation baseline
3. Significant macroeconomic changes occur (recession, policy changes)
4. New data becomes available that changes feature-target relationships

### Recalibration Methods

| Method | Description | When Used |
|--------|-------------|-----------|
| Platt Scaling | Logistic regression on scores | Mild calibration drift |
| Isotonic Regression | Non-parametric recalibration | Non-linear calibration drift |
| Full Retraining | Complete model rebuild | Significant performance degradation |

## 14.6 Fairness and Bias Considerations

Credit risk models must be tested for discriminatory bias across protected characteristics:

| Protected Class | Analysis | Acceptable Threshold |
|----------------|----------|---------------------|
| Gender | AUC comparison across gender groups | AUC difference < 3% |
| Age | Performance comparison across age bands | AUC difference < 5% |
| Geography | Equal treatment across provinces | Approval rate difference < 10% |

Bias mitigation techniques include:
- Adversarial debiasing during training
- Threshold optimization by subgroup
- Post-processing calibration to equalize outcomes

## 14.7 Why Black-Box Models Are Risky in Finance

Black-box models present specific risks in financial services:

| Risk Category | Description | Mitigation |
|--------------|-------------|------------|
| Regulatory Rejection | Regulators may refuse to approve unexplainable models | SHAP analysis, surrogate models |
| Fair Lending Violations | Inability to prove non-discrimination | Bias testing, disparate impact analysis |
| Model Misuse | Users apply model outside intended scope | Clear documentation, use-case boundaries |
| Debugging Difficulty | Hard to diagnose failures | Feature importance, error analysis |
| Legal Challenge | Adverse decisions cannot be explained | Individual SHAP explanations |

This solution addresses these concerns through comprehensive SHAP-based explainability, enabling both global model understanding and individual decision explanations.

# 15. Stress Testing and Scenario Analysis

## 15.1 Introduction to Stress Testing

Stress testing is a risk management technique that evaluates how a portfolio performs under extreme but plausible adverse conditions. Regulatory requirements (CCAR, EBA stress tests) mandate periodic stress testing of credit portfolios.

### Relationship to IFRS 9

While IFRS 9 requires forward-looking ECL estimates, it does not mandate specific stress scenarios. However, stress testing provides:
- **Scenario calibration:** Validates the range of forward-looking adjustments
- **Capital adequacy assessment:** Ensures sufficient buffers under adverse conditions
- **Risk appetite validation:** Confirms that portfolio risk remains within limits

## 15.2 Stress Scenario Design

### Macroeconomic Scenarios

| Scenario | Description | Default Rate Impact |
|----------|-------------|-------------------|
| Baseline | Expected economic conditions | No adjustment |
| Adverse | Mild recession, unemployment +3% | PD multiplier: 1.5x |
| Severely Adverse | Severe recession, unemployment +6% | PD multiplier: 2.5x |
| Extreme | Financial crisis conditions | PD multiplier: 4.0x |

### Scenario Parameters

| Variable | Baseline | Adverse | Severe | Extreme |
|----------|----------|---------|--------|---------|
| GDP Growth | 2.5% | -1.0% | -3.5% | -6.0% |
| Unemployment | 5.0% | 8.0% | 11.0% | 15.0% |
| Interest Rates | Current | +200bps | +400bps | +600bps |
| Inflation | 2.0% | 5.0% | 8.0% | 12.0% |

### Stress Testing Methodology

The stress testing approach applies multipliers to the model's predicted probabilities:

$$\text{PD}_{stressed} = \min(\text{PD}_{base} \times \text{Multiplier}, 0.99)$$

The cap at 99% ensures numerical stability while allowing near-certainty for the highest-risk loans.

### Stressed ECL Calculation

Under stress scenarios, ECL increases due to:
1. Higher PDs (direct model output adjustment)
2. Potentially higher LGDs (collateral value decline)
3. Potential EAD increase (payment holidays, forbearance)

$$\text{ECL}_{stressed} = \text{PD}_{stressed} \times \text{LGD}_{stressed} \times \text{EAD}_{stressed}$$

## 15.3 Portfolio Resilience Interpretation

The stress test results indicate portfolio vulnerability:

| Resilience Level | Provision Increase Under Adverse | Interpretation |
|-----------------|----------------------------------|----------------|
| Strong | < 50% | Well-positioned for downturns |
| Moderate | 50-100% | Adequate but monitor concentrations |
| Weak | 100-200% | Vulnerable; risk reduction recommended |
| Critical | > 200% | Immediate action required |

## 15.4 Sensitivity Analysis

Sensitivity analysis examines how changes in individual assumptions affect ECL:

| Assumption Change | ECL Impact | Interpretation |
|------------------|------------|----------------|
| LGD +10% | Linear increase in ECL | Collateral quality is key driver |
| PD +20% | Linear increase in ECL | Model sensitivity to economic conditions |
| Correlation +0.1 | Super-linear increase | Portfolio concentration risk |

The sensitivity analysis identifies which assumptions have the greatest impact on provisions, guiding where risk management should focus its attention.


```python
# ============================================================
# STRESS TESTING & SCENARIO ANALYSIS
# ============================================================

# Define stress scenarios
scenarios = {
    'Baseline': {'pd_multiplier': 1.0, 'lgd_multiplier': 1.0, 'description': 'Expected economic conditions'},
    'Adverse': {'pd_multiplier': 1.5, 'lgd_multiplier': 1.1, 'description': 'Mild recession, unemployment +3%'},
    'Severe': {'pd_multiplier': 2.5, 'lgd_multiplier': 1.25, 'description': 'Severe recession, unemployment +6%'},
    'Extreme': {'pd_multiplier': 4.0, 'lgd_multiplier': 1.5, 'description': 'Financial crisis conditions'},
}

stress_results = []

print("=" * 70)
print("STRESS TESTING RESULTS")
print("=" * 70)
print(f"{'Scenario':<12} {'PD Mult':>8} {'LGD Mult':>8} {'Avg PD':>8} {'Total ECL':>12} {'ECL Increase':>12}")
print("-" * 70)

baseline_ecl = None

for scenario_name, params in scenarios.items():
    # Apply stress to the portfolio
    stressed_pd = np.clip(staging_df['pd_12m'] * params['pd_multiplier'], 0, 0.99)
    stressed_lgd = LGD_ASSUMPTION * params['lgd_multiplier']
    stressed_ecl = stressed_pd * stressed_lgd * staging_df['ead']

    total_ecl_stressed = stressed_ecl.sum()
    avg_pd_stressed = stressed_pd.mean()

    if baseline_ecl is None:
        baseline_ecl = total_ecl_stressed
        ecl_increase = "—"
    else:
        increase_pct = (total_ecl_stressed / baseline_ecl - 1) * 100
        ecl_increase = f"+{increase_pct:.1f}%"

    print(f"{scenario_name:<12} {params['pd_multiplier']:>8.1f} {params['lgd_multiplier']:>8.2f} "
          f"{avg_pd_stressed:>7.2%} ${total_ecl_stressed:>10,.0f} {ecl_increase:>12}")

    stress_results.append({
        'Scenario': scenario_name,
        'PD_Multiplier': params['pd_multiplier'],
        'LGD_Multiplier': params['lgd_multiplier'],
        'Avg_PD': avg_pd_stressed,
        'Total_ECL': total_ecl_stressed,
        'ECL_Increase_Pct': (total_ecl_stressed / baseline_ecl - 1) * 100 if baseline_ecl else 0
    })

stress_df = pd.DataFrame(stress_results)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ECL by scenario
bars = axes[0].bar(stress_df['Scenario'], stress_df['Total_ECL'], 
                   color=[COLORS['success'], '#F39C12', '#E67E22', COLORS['danger']],
                   edgecolor='white')
axes[0].set_ylabel('Total Expected Credit Loss ($)')
axes[0].set_title('Stressed ECL by Scenario', fontweight='bold')
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'${y/1e6:.1f}M'))

# Add value labels
for bar, val in zip(bars, stress_df['Total_ECL']):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + total_ecl*0.02,
                f'${val/1e6:.2f}M', ha='center', va='bottom', fontweight='bold')

# ECL increase percentage
increase_data = stress_df[stress_df['Scenario'] != 'Baseline']
axes[1].bar(increase_data['Scenario'], increase_data['ECL_Increase_Pct'],
            color=['#F39C12', '#E67E22', COLORS['danger']], edgecolor='white')
axes[1].set_ylabel('ECL Increase from Baseline (%)')
axes[1].set_title('Provision Impact Under Stress', fontweight='bold')
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0f}%'))

# Add value labels
for i, (scenario, val) in enumerate(zip(increase_data['Scenario'], increase_data['ECL_Increase_Pct'])):
    axes[1].text(i+1, val + max(stress_df['ECL_Increase_Pct'])*0.02,
                f'+{val:.0f}%', ha='center', va='bottom', fontweight='bold')

plt.suptitle('Credit Portfolio Stress Testing Analysis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Stage migration under stress
print("\n" + "=" * 70)
print("STAGE MIGRATION UNDER SEVERE STRESS SCENARIO")
print("=" * 70)

stressed_severe_pd = np.clip(staging_df['pd_12m'] * 2.5, 0, 0.99)
staging_severe = staging_df.copy()
staging_severe['stressed_pd'] = stressed_severe_pd
staging_severe['stressed_stage'] = staging_severe['stressed_pd'].apply(assign_stage)

migration = pd.crosstab(staging_df['stage'], staging_severe['stressed_stage'], 
                        margins=True, normalize='index')
print(migration.to_string())

# Sensitivity Analysis
print("\n" + "=" * 70)
print("SENSITIVITY ANALYSIS: PD MULTIPLIER IMPACT")
print("=" * 70)
sensitivity_multipliers = [1.0, 1.25, 1.5, 1.75, 2.0, 2.5, 3.0, 4.0]
sensitivity_results = []

for mult in sensitivity_multipliers:
    sens_pd = np.clip(staging_df['pd_12m'] * mult, 0, 0.99)
    sens_ecl = (sens_pd * LGD_ASSUMPTION * staging_df['ead']).sum()
    sensitivity_results.append({'PD_Multiplier': mult, 'ECL': sens_ecl})

sens_df = pd.DataFrame(sensitivity_results)
print(sens_df.to_string(index=False))

# Plot sensitivity curve
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(sens_df['PD_Multiplier'], sens_df['ECL'] / 1e6, 'o-', color=COLORS['primary'], linewidth=2, markersize=8)
ax.set_xlabel('PD Stress Multiplier')
ax.set_ylabel('Total ECL ($ Millions)')
ax.set_title('ECL Sensitivity to PD Multiplier', fontweight='bold')
ax.axvline(x=1.0, color='gray', linestyle='--', alpha=0.7, label='Baseline')
ax.axvline(x=2.5, color=COLORS['danger'], linestyle='--', alpha=0.7, label='Severe Scenario')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
